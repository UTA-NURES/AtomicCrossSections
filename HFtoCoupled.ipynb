```python
from sympy.physics.quantum.cg import CG
from sympy import S
import numpy as np
```

```python
j1, j2 = 1, 1
m_vals = [-1, 0, 1]
m_to_state = {-1: "c", 0: "b", 1: "d"}

for m1 in m_vals:
    for m2 in m_vals:
        print(f"\n|{m_to_state[m1]}> ⊗ |{m_to_state[m2]}> =")
        found = False
        for F in [0, 1, 2]:  # Allowed total angular momentum values
            for MF in range(-F, F+1):
                cg = CG(S(j1), S(m1), S(j2), S(m2), S(F), S(MF)).doit()
                if cg != 0:
                    found = True
                    print(f"  {cg} * |F={F}, mF={MF}>")

state = "a"
j1, j2, m2 = 1, 0, 0
m_vals = [-1, 0, 1]
for m1 in m_vals:
    print(f"\n|{m_to_state[m1]}> ⊗ |{state}> =")
    found = False
    for F in [0, 1, 2]:  # Allowed total angular momentum values
        for MF in range(-F,F+1):
            cg = CG(S(j1), S(m1), S(j2), S(m2), S(F), S(MF)).doit()
            if cg != 0:
                found = True
                print(f"  {cg} * |F={F}, mF={MF}>")
                
j1, j2, m1 = 0, 1, 0
m_vals = [-1, 0, 1]
for m2 in m_vals:
    print(f"\n|{state}> ⊗ |{m_to_state[m2]}> =")
    found = False
    for F in [0, 1, 2]:  # Allowed total angular momentum values
        for MF in range(-F,F+1):
            cg = CG(S(j1), S(m1), S(j2), S(m2), S(F), S(MF)).doit()
            if cg != 0:
                found = True
                print(f"  {cg} * |F={F}, mF={MF}>")
                
j1, m1 = 0, 0
j2, m2 = 0, 0
M = m1 + m2

for J in [0, 1, 2]:  # 1 ⊗ 1 = 0 ⊕ 1 ⊕ 2
    cg = CG(S(j1), S(m1), S(j2), S(m2), S(J), S(M)).doit()
    if cg != 0:
        print(f"|a> ⊗ |a> = {cg} * |F={J}, mF={M}>")
```
